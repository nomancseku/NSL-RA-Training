```python
A = [5,0,9,2,7,3,4]

MAX_N = 20
EMPTY_VALUE = -1

mem = list()

def f(i, A):
    if mem[i] != EMPTY_VALUE:
        return mem[i]
    
    ans = 0
    l = len(A)
    for j in range(i+1, l):
        if A[j]>A[i]:
            ans = max(ans, f(j,A))
            
    mem[i] = ans + 1
    return mem[i]
    
def find_LIS(A = A):
    ans = 0
    l = len(A)
    for i in range(l):
        mem.append(EMPTY_VALUE)
    
    for i in range(l):
        ans = max(ans, f(i,A))
    
    return ans
```

```python
print(find_LIS(A = A))
```
